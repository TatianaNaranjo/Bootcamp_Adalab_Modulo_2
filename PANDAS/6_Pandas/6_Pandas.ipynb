# Ejercicio 1
쮺u치ntos ataques hubo por pa칤s? Devuelve un dataframe.
游늷 NOTA Ordena los resultados de mayor a menor. Esto todav칤a no lo hemos visto, pero un dataframe lo podemos ordenar con .sort_values(). Aqu칤 ten칠is una explicaci칩n de como hacerlo.


```python
df = pd.read_csv("union_ataques.csv")
```

```python
df_groupby_paises = pd.DataFrame(df.groupby("Country")["Case Number"].count())
df_groupby_paises.rename(columns = {'Case Number':'Attacks Number'}, inplace = True)
df_groupby_paises_or= df_groupby_paises.sort_values(by='Attacks Number', ascending=False)
resultado_paises_ord = df_groupby_paises_or.reset_index()
resultado_paises_ord 
```

# Ejercicio 2 
쮺u치ntos ataques ha habido por a침o? Devuelve un dataframe. De nuevo ordena los resultados de mayor a menor.

```python
df_groupby_year = pd.DataFrame(df.groupby("Year")["Case Number"].count())
df_groupby_year.rename(columns = {'Case Number':'Attacks Number'}, inplace = True)
df_groupby_year_or= df_groupby_year.sort_values(by='Attacks Number', ascending=False)
resultado_group_year_or = df_groupby_year_or.reset_index()
resultado_group_year_or 
```
